Now, we need to import modules from CRYSTALClear:

```python
from CRYSTALClear.crystal_io import External_unit
from CRYSTALClear import plot as CCplt
```

## Phonon Bands
Specify your output file path:

```python
path_bands = 'outputs/diamond_bands.f25'
```

Let's create a CRYSTALClear object:

```python
data = External_unit().read_phonon_band(path_bands)
CCplt.plot_phonon_band(data, energy_range=[0,1500], k_labels=['Gamma', 'W', 'X', 'Gamma','L'], title='Diamond', color='#AB2929')
```

## Phonon Density-of-States
Specify your output file path:

```python
path_dos = 'outputs/diamond_pdos.f25'
```

Again create a CRYSTALClear object:

```python
data = External_unit().read_phonon_dos(path_dos)
CCplt.plot_phonon_dos(data, freq_range=[0,1400], figsize=(10,4), title='Diamond')
```
